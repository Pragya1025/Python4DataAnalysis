```python
import pandas as pd
from datetime import datetime
dfplayer=pd.read_csv('C:\Python35\Scripts\data\Player.csv')
dfplayerattribute=pd.read_csv('C:\Python35\Scripts\data\Player_Attributes.csv')
one=pd.merge(dfplayer, dfplayerattribute, on='player_api_id', how='right')


analysis1_D = one[['player_name','date','preferred_foot','overall_rating']]

analysis1_D = analysis1_D.loc[(analysis1_D['date'].str.startswith("2014") ) |(analysis1_D['date'].str.startswith("2013"))]


analysis1_D.loc[analysis1_D['date'].str.startswith("2014"), 'date']='2014'

analysis1_D.loc[analysis1_D['date'].str.startswith("2013"), 'date']='2013'

year = ['2013', '2014']

a = analysis1_D[analysis1_D['preferred_foot'] == 'left' ].groupby(['player_name', 'preferred_foot']).max()
b = analysis1_D[analysis1_D['preferred_foot'] == 'right'].groupby(['player_name', 'preferred_foot']).max()

for i in year:

    print(i)
    c = a[a['date']==str(i)]
    d = b[b['date'] == str(i)]

    o=c.sort_values(['overall_rating'], ascending=[False]).head(8)
    p=d.sort_values(['overall_rating'], ascending=[False]).head(8)
    frames = [o,p]

    result=pd.concat(frames)

    print(result)
    result.to_csv('output1.xls', sep='\t', encoding='utf-8') 
```
