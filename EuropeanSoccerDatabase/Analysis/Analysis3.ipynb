```python
# Analysis 3
# FOR EPL, get the  Team Attributes for each team over the seasons present in the Match Attribute

import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from numpy import zeros, arange, array


dfLeague = pd.read_csv('C:\Python35\Scripts\data\League.csv')
EPLID = dfLeague.loc[dfLeague["name"]=='England Premier League', "country_id"]
EPLID =  EPLID.values

print("EPLID", EPLID[0])


dfMatch = pd.read_csv('C:\Python35\Scripts\data\Match.csv')


dfMatch = dfMatch.loc[dfMatch["league_id"] == EPLID[0]]

dfMatch = dfMatch[["season", "home_team_api_id"]]


matchTeamId = dfMatch["home_team_api_id"].unique()
print("UNIQUE TEAM ID")
print(matchTeamId)


dfTeam=pd.read_csv('C:\Python35\Scripts\data\Team.csv')
dfTeamAttributes=pd.read_csv('C:\Python35\Scripts\data\Team_Attributes.csv')
mergeData = pd.merge(dfTeam, dfTeamAttributes, on='team_api_id', how='right')

mergeData = mergeData[["team_api_id", "date", "team_long_name", "buildUpPlaySpeed", "chanceCreationShooting", "defencePressure"]]



mergeData.loc[mergeData["date"].str.startswith("2015"), 'date']='2015'
mergeData.loc[mergeData["date"].str.startswith("2014"), 'date']='2014'
mergeData.loc[mergeData['date'].str.startswith("2013"), 'date']='2013'
mergeData.loc[mergeData['date'].str.startswith("2012"), 'date']='2012'
mergeData.loc[mergeData['date'].str.startswith("2011"), 'date']='2011'
mergeData.loc[mergeData['date'].str.startswith("2010"), 'date']='2010'

mergeData.to_csv('output3.xls', sep='\t', encoding='utf-8') 

def plotGraph(plotdata):
    print (len(plotdata))
    speed = zeros(len(plotdata))
    shooting = zeros(len(plotdata))
    defence = zeros(len(plotdata))
    x = zeros(len(plotdata))
    i = 0
    for row in plotdata.iterrows():
        teamname = row[1].team_long_name
        speed[i] = row[1].buildUpPlaySpeed
        shooting[i] = row[1].chanceCreationShooting
        defence[i] = row[1].defencePressure
        x[i] = row[1].date
        i+=1

    speed = tuple(speed)
    shooting = tuple(shooting)
    defence = tuple(defence)
    x = tuple(x)

    N = 6
    ind = array([0, 2, 4, 6, 8, 10])  # the x locations for the groups
    width = 0.45  # the width of the bars
    # w = 0.35
    fig, ax = plt.subplots()
    rects1 = ax.bar(ind, speed, width, color='r')
    rects2 = ax.bar(ind + width, shooting, width, color='y')
    rects3 = ax.bar(ind + (width *2), defence, width, color='g')
    # add some text for labels, title and axes ticks
    ax.set_ylabel('Range')
    stat = teamname + '  Stats'
    ax.set_title(stat)
    ax.set_xticks(ind + width)
    ax.set_xticklabels(x)


    ax.legend((rects1[0], rects2[0], rects3[0]), ('Speed', 'Shooting', 'Defense'))

    # plt.show()
    plt.savefig('output3/'+str(stat)+'.png')


    print(teamname)
    print(speed)
    print(shooting)
    print(defence)
    print(x)



for i in matchTeamId:
    print(i)
    EPLTeamData = mergeData.loc[mergeData["team_api_id"] == i]
    EPLTeamData.groupby('date').mean()

    print("MEAN EPL DATA")
    print(EPLTeamData)
    plotGraph(EPLTeamData)




```
