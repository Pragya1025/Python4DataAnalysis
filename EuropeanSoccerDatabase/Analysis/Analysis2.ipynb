```python
# Analysis 2
# country or league grouped by season and stage and den compare home team goal and away team goal

import pandas as pd
from numpy import sum, zeros
import matplotlib.pyplot as plt

dfleague=pd.read_csv('C:\Python35\Scripts\data\league.csv')
dfmatch=pd.read_csv('C:\Python35\Scripts\data\match.csv')

two=pd.merge(dfleague, dfmatch, on='country_id', how='right')

# print(two)
two.isnull().values.any()
two.fillna(method='ffill')

df = two[['name','season','home_team_goal','away_team_goal']]


df1=df.groupby(['season', 'name']).agg(sum).reset_index()
print(df1)
df1.to_csv('output2.xls', sep='\t', encoding='utf-8') 
def plotGraph(data, year):
    size = len(data)

    n = list()
    home_goal = zeros(size)
    away_goal = zeros(size)

    i = 0;
    for row in data.iterrows():
        # teamname = row[1].team_long_name
        print(row[1])
        print(row[1].teamName)
        name = str(row[1].teamName)
        sep = ' '
        name = name.split(sep, 1)[0]
        n.append(name)
        home_goal[i] = row[1].home_team_goal
        away_goal[i] = row[1].away_team_goal
        i+=1

    print(n)
    print(home_goal)
    print(away_goal)

    colors = ['yellow','green','red', 'black', 'grey', 'pink','purple', 'silver' ,'gold', 'lightskyblue', 'lightcoral']
    plt.subplot(211)
    patches, texts = plt.pie(home_goal, colors=colors, shadow=True, startangle=90)
    plt.legend(patches, n, loc="best")
    plt.axis('equal')
    plt.tight_layout()
    plt.title(year+" Home Goal")

    plt.subplot(212)
    patches, texts = plt.pie(away_goal, colors=colors, shadow=True, startangle=90)
    # plt.legend(patches, n, loc="best")
    plt.axis('equal')
    plt.tight_layout()
    plt.title(year + " Away Goal")
    plt.savefig('output2/'+str(year)+'.png')

    #plt.show()


year = ['2008/2009', '2009/2010', '2010/2011', '2011/2012', '2012/2013', '2013/2014', '2014/2015'
        ,'2015/2016']

for i in year:
    a = df1[df1["season"] == i]
    a = a.rename(columns={'name': 'teamName'})
    print(a)
    i = str(i)
    i = i.replace('/','-')
    plotGraph(a, i)
```
