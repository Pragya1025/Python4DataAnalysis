```python
#Analysis 4
# Create folder for output4
# in LA LIGA, STAT OF THE EACH TEAM OVER THE PERIOD OF YEARS

import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from numpy import zeros, arange, array, where

dfLeague = pd.read_csv('C:\Python35\Scripts\data\League.csv')
SpanishLeague = dfLeague.loc[dfLeague["name"]=='Spain LIGA BBVA', "country_id"]
SpanishLeague =  SpanishLeague.values

print(SpanishLeague)

dfMatch = pd.read_csv('C:\Python35\Scripts\data\Match.csv')
dfMatch = dfMatch.loc[dfMatch["league_id"] == SpanishLeague[0]]

dfMatch = dfMatch[["season", "stage", "home_team_api_id", "home_team_goal", "away_team_goal"]]

print("MATCH")
print(dfMatch)

matchTeamId = dfMatch["home_team_api_id"].unique()
print("UNIQUE TEAM ID")
print(matchTeamId)

dfTeam=pd.read_csv('C:\Python35\Scripts\data\Team.csv')
dfTeam = dfTeam[["team_api_id", "team_long_name"]]
# print(dfTeam)

dfTeam.to_csv('output4.xls', sep='\t', encoding='utf-8') 
TeamName = list()
j=0
for i in matchTeamId:
    d =  dfTeam.loc[dfTeam["team_api_id"]==i, "team_long_name"]
    print(d.values)
    d = d.values
    TeamName.append(d[0])
    j+=1



def plotGraph(teamname, winGame, lostGame, drawGame, year):


    winGame = tuple(winGame)
    lostGame = tuple(lostGame)
    drawGame = tuple(drawGame)
    year = tuple(year)


    N = 6
    ind = array([0, 2, 4, 6, 8, 10])  # the x locations for the groups
    width = 0.45  # the width of the bars
    # w = 0.35
    fig, ax = plt.subplots()
    rects1 = ax.bar(ind, winGame, width, color='r')
    rects2 = ax.bar(ind + width, lostGame, width, color='y')
    rects3 = ax.bar(ind + (width *2), drawGame, width, color='g')
    # add some text for labels, title and axes ticks
    ax.set_ylabel('Matches')
    stat = teamname + ' Match  Stats'
    ax.set_title(stat)
    ax.set_xticks(ind + width)
    ax.set_xticklabels(year)


    ax.legend((rects1[0], rects2[0], rects3[0]), ('Win', 'Lost', 'Draw'))

    # plt.show()
    plt.savefig('output4/'+str(stat)+'.png')


teamIndex = 0
for i in matchTeamId:
    newData = dfMatch.loc[dfMatch["home_team_api_id"] == i]
    print(newData)
    year = ['2010', '2011', '2012', '2013', '2014', '2015']

    win = zeros(len(year))
    lost = zeros(len(year))
    draw = zeros(len(year))
    k=0
    for j in year:
        print(len(newData[(newData['season'].str.startswith(j)) &(newData['home_team_goal'] > newData['away_team_goal'])]))
        win[k] = len(newData[(newData['season'].str.startswith(j)) &(newData['home_team_goal'] > newData['away_team_goal'])])
        lost[k] = len(newData[(newData['season'].str.startswith(j)) &(newData['home_team_goal'] < newData['away_team_goal'])])
        draw[k] = len(newData[(newData['season'].str.startswith(j)) & (newData['home_team_goal'] < newData['away_team_goal'])])
        k+=1

    print(win)
    print(lost)
    print(draw)

    plotGraph(TeamName[teamIndex], win, lost, draw, year)
    teamIndex+=1


```
